Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path='Sarcasm_Headlines_Dataset.json'
df=pd.read_json(file_path, lines=True)
print(df.head())
```

```python
df['headline']
```

```python
df['is_sarcastic']
```

```python
import re
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer


nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')

# Define outside the loop for performance
lemmatizer = WordNetLemmatizer()
custom_stopwords = set(stopwords.words('english')) - {'no', 'not', 'never'}

corpus = []

for i in range(0, len(df)):
    # omitting non-alphabetic characters
    headline = re.sub('[^a-zA-Z]', ' ', df['headline'][i])
    headline = headline.lower()
    tokens = word_tokenize(headline)

    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    tokens = [word for word in tokens if word not in custom_stopwords]

    # joining into cleaned string
    cleaned_headline = ' '.join(tokens)
    corpus.append(cleaned_headline)
```

```python
print(corpus)
```

creating bag of words model

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
X=cv.fit_transform(corpus).toarray()
y=df.iloc[:, -1].values #taking the last column as 'y'
```

```python
len(X[0])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_pred=classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score=accuracy_score(y_test, y_pred)
print(accuracy_score)
```
